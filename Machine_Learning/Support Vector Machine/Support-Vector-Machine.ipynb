```python
#importing library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data-set

d=pd.read_csv('Social_Network_Ads.csv')
print(d)
print(f'size of data = {d.shape}')
```

```python
#Let's split the data-set into dependent(y) and independent(x)

x=d.iloc[:,0:2].values
y=d.iloc[:,2:3].values
```

```python
#let's split data into train & test data 

from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#Let's apply feature scalling

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtr=sc.fit_transform(xtr)
xts=sc.transform(xts)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear',random_state=0)
svc.fit(xtr,ytr)
```

```python
y_predict=svc.predict(xts)
print(y_predict)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(yts,y_predict)
score=accuracy_score(yts,y_predict)


print(cm)
print(score)
```
