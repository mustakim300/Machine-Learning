# Logistic Reression

Logistic regression is named for the function used at the core of the method, the logistic function. The logistic function or the sigmoid function is an S-shaped curve that can take any real-valued number and map it into a value between 0 and 1.

```python
from IPython.display import Image
Image("Logistic_Reression.png")
```

```python
#Importing Libraries


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import Data-set

d=pd.read_csv('Social_Network_Ads.csv')
print(d)
```

```python
#lets divied the data into dependent and independent data


x=d.iloc[: ,:-1].values    #it will take Age and EstimatedSalary values
y=d.iloc[: ,2:3].values    #it will take Purchased values
```

```python
#let's devide the data into train and test 


from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#feature scaling


from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtr=sc.fit_transform(xtr)
xts=sc.transform(xts)
```

```python
# importing model


from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(random_state = 0)
LR.fit(xtr,ytr)
```

```python
#let's check predict values

y_predict = LR.predict(xts)
print('[Predict value || Test value]')
print(np.concatenate((y_predict.reshape(len(y_predict),1), yts.reshape(len(yts),1)),1))
```
