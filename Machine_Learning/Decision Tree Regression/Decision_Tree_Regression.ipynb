# Decision Tree Regression

What is Decision tree?

Decision tree builds regression or classification models in the form of a tree structure. It breaks down a dataset into smaller and smaller subsets while at the same time an associated decision tree is incrementally developed. The final result is a tree with decision nodes and leaf nodes

```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset

d=pd.read_csv('Position_Salaries.csv')
```

```python
#check the first five rows of dataset

print(d.head())
```

```python
#Let's split data into x and y

x=d.iloc[:,1:2].values
y=d.iloc[:,2:3].values
print(f'values of x are :-\n {x}')
print(f'values of y are :-\n {y}')
```

```python
plt.scatter(x,y,label=('Original Data'))
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
DTR=DecisionTreeRegressor()
DTR.fit(x,y)
```

```python
d
```

```python
y_predict=DTR.predict(x)
y_predict
```

```python
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))

plt.scatter(x,y,label=('Original Data'))
plt.plot(x_grid,DTR.predict(x_grid),label=('DTR'),c='red')

plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
#let's predict the salary for a level
n=input()
y_predict=DTR.predict([[n]])
print(f'Salary is:- {y_predict}')
```
